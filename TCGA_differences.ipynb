```python
def getAllColumns(columns_list, dataFrame):
    for column in dataFrame.columns:

        if not column in columns_list:
                columns_list.append(column)

        for elems in dataFrame[column]:

            if isinstance(elems, dict):
                for key, value in elems.items():
                    if isinstance(value, dict):
                        getAllColumns(columns_list, pd.DataFrame([value]))
                    elif not column in columns_list:
                        columns_list.append(column)

            elif isinstance(elems, list):
                getAllColumns(columns_list, pd.DataFrame(elems))
```
